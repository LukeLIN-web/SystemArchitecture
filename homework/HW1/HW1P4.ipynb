First, we know the probability density function of the Poisson distribution. Then, by accumulating, we can obtain the cumulative distribution function.

We can use Inverse Cumulative Function Method: The cumulative distribution function of a Poisson distribution is a non-decreasing piecewise function. 
If we consider this function as a random variable, then this random variable follows a uniform distribution on the interval [0, 1].
Iteration is used to determine which segment of the cumulative distribution function the current random value falls into.

refer: 

1. https://blogs.sas.com/content/iml/2013/07/22/the-inverse-cdf-method.html
2. Ross, Sheldon M. Simulation https://picx.zhimg.com/80/08b54b698f8be574c3e5a4f70dce1441_1440w.webp?source=1940ef5c 

```python
import math
import random
import time
import matplotlib.pyplot as plt

def poisson(lambda_value):
    x = 0
    p = math.exp(-lambda_value)
    s = p
    u = random.random()

    while u > s:
        x += 1
        p = p * lambda_value / x # from the poisson formula
        s += p

    return x

def main():
    random.seed(time.time())

    mean = 5.0  # lambda
    num_samples = 1000

    values = []
    for i in range(num_samples):
        value = poisson(mean)
        values.append(value)

    plt.hist(values, bins=max(values)-min(values)+1, rwidth=0.8, align='left', density=True)
    plt.xlabel("Value")
    plt.ylabel("Frequency (Normalized)")
    plt.title("Normalized Value Frequency Histogram")
    plt.grid(axis='y')
    plt.show()

if __name__ == "__main__":
    main()
```
